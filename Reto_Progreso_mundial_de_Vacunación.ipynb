```python
#Extraer información del archivo
import pandas as pd
vacc=pd.read_csv("country_vaccinations.csv")
vacc
```

```python
#b)Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64
vacc["date"]=pd.to_datetime(vacc["date"])
vacc.info()
```

```python
#Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).
vacc[["daily_vaccinations","vaccines"]].groupby("vaccines").sum()
```

```python
#d. Obtener la cantidad de vacunas aplicadas en todo el mundo.
#metodo 1
sum1=vacc["daily_vaccinations"].sum()
sum1
#metodo2
maxvacc=vacc[["total_vaccinations","country"]].groupby("country").max()
sumamax=maxvacc["total_vaccinations"].sum()
sumamax
```

```python
#e. Calcular el promedio de vacunas aplicadas por país.
vacc[["daily_vaccinations","country"]].groupby("country").mean()
```

```python
#f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
fecha=vacc.query('date=="29/01/21"')
suma_total = fecha["daily_vaccinations"].sum()
print("Suma total de la columna 'total':", suma_total)
```
